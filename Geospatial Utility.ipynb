```python
plot = GMapPlot(
    x_range=x_range, y_range=y_range,
    map_options=map_options,
    api_key=API_KEY,
)

print("Initializing Plot...")
```

```python
testdf = pd.read_csv('./ECE180-Team2/python_scraper/geospatialData/FA16-geospatial.csv',sep='\s*,\s*',engine = 'python')
lat_list = testdf['longitude'].values.tolist()
lon_list = testdf['latitude'].values.tolist()
dot_size = testdf['students_number'].values.tolist()

print("Reading from csv file...")
```

```python
DATA1 = {'lat':lat_list[:50], 'lon':lon_list[:50]}
source1 = ColumnDataSource(
    data=DATA1
    )

DATA2 = {'lat':lat_list[50:], 'lon':lon_list[50:]}
source2 = ColumnDataSource(
    data=DATA2
    )

print("Constructing data objects...")
```

```python
circle1 = Circle(x="lon", y="lat", size=15, fill_color="blue", line_color="black")
plot.add_glyph(source1, circle1)

circle2 = Circle(x="lon", y="lat", size=10, fill_color="red", line_color="black")
plot.add_glyph(source2, circle2)

print("Adding data points to plot...")
```

```python
pan = PanTool()
wheel_zoom = WheelZoomTool()
box_select = BoxSelectTool()

plot.add_tools(pan, wheel_zoom, box_select)

doc = Document()
doc.add_root(plot)

if __name__ == "__main__":
    filename = "maps.html"
    with open(filename, "w") as f:
        f.write(file_html(doc, INLINE, "Google Maps Example"))
    print("Wrote %s" % filename)
    view(filename)
```
